# parkinson's disease prediction

```python
#importing the necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

# Data collection

```python
#load the dataset
diabetes_dataset = pd.read_excel('/content/Parkinsson disease.xlsx')
diabetes_dataset.head()
```

```python
# getting some information about the dataset
diabetes_dataset.info()
```

```python
#statestical measures
diabetes_dataset.describe()
```

```python
#distribution of target and feature variable
diabetes_dataset['status'].value_counts()
```

1  -> parkinson's positive
0 -> healthy

```python
diabetes_dataset=diabetes_dataset.drop(['name'],axis=1)
diabetes_dataset.groupby('status').mean()
```

```python
x=diabetes_dataset.drop(columns='status',axis=1)
y=diabetes_dataset['status']
```

```python
print(x)
print(y)
```

#training and testing data splitting

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
print(x_train.shape,x_test.shape,x.shape)
```

#standardize our data

```python
scalar=StandardScaler()
scalar.fit(x_train)
x_train=scalar.transform(x_train)
x_test=scalar.transform(x_test)
model=svm.SVC(kernel='linear')
model.fit(x_train,y_train)
```

#model evaluation

```python
x_train_pred=model.predict(x_train)
train_accur=accuracy_score(y_train,x_train_pred)
print(train_accur)
```

```python
x_test_pred=model.predict(x_test)
test_accur=accuracy_score(y_test,x_test_pred)
print(test_accur)
```
